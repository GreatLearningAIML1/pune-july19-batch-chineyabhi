```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('Laliga.csv')
dataset.head()
```

```python
dataset.columns = dataset.loc[0,:] #setting column names correctly as they are named as Unnamed at this moment
dataset.head()
```

### 1.Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data. (5 points)

```python

dataset.replace('-',0,inplace = True)
dataset.head(10) #verification that all - is changed to 0
```

### 2.Print all the teams which have started playing between 1930-1980. (5 points)

```python
#removing first row as it is string (same as column names)
dataset.drop(dataset.index[0], inplace = True)
```

```python
dataset.dtypes #checking for datatypes of columns
```

```python
#saving the required Debut in an array for operation
strings = np.array(dataset.Debut)
list_debut_year = []
strings
```

```python
#changing year format 19xx-yy to 19xx and storing it in a separate list 
for x in strings:
    debut_year = x[:4]
    list_debut_year.append(debut_year)
```

```python
#replacing the 19xx-yy formatted column by new list which has the same years as 19xx format
dataset['Debut']= list_debut_year
```

```python
dataset.head(10) #verification that replacement is successful 
```

```python
dataset['Debut'] = dataset['Debut'].astype(int) #converting type to int of the column to perform arithmetic operations
```

```python

#setting the filter of debut from 1930-1980 and saving that in a new dataframe and printing that
df3 = dataset.loc[(dataset.Debut>=1930) & (dataset.Debut<=1980),['Team','Debut']].sort_values('Debut',ascending = True)
print('All the teams which have started playing between 1930-1980')
df3
```

### 4.Write a function with name Goal_diff_count which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points)

```python
#Function to return all the teams with their Goal Differences
#It takes a data frame as argument and returns a data frame which contains Team and corresponding Goal Difference
def Goal_diff_count(df):
    goal_diff = pd.DataFrame({'Team':df['Team'],'Goal Diff':df['GoalsFor'] - df['GoalsAgainst']})
    return goal_diff

#converting type to int of the column to perform arithmetic operations
dataset['GoalsFor'] = dataset['GoalsFor'].astype(int)
dataset['GoalsAgainst'] = dataset['GoalsAgainst'].astype(int)

#Function call
minmax = Goal_diff_count(dataset)

#Output displays
print('All the teams with their Goal Differences \n')
print(minmax)

```

```python
#Min/max Goal Difference teams 
print("Team with highest goal difference =", minmax.loc[minmax['Goal Diff'].idxmax(axis = 0),'Team'])
print("Team with lowest goal difference =", minmax.loc[minmax['Goal Diff'].idxmin(axis = 0),'Team'])
#Achieved both requirements in question (all teams with GD and max/min GD teams) with just one call of the function in the previous cell
```

### 5.Create a new column with name Winning Percent and append it to the data set (5 points)

```python
#changing data type to perform arithmetic operations
dataset['GamesPlayed'] = dataset['GamesPlayed'].astype(int)
dataset['GamesWon'] = dataset['GamesWon'].astype(int)

#calculating win% and adding it to new column 'Winning Percent'
dataset['Winning Percent'] = (dataset['GamesWon']*100)/dataset['GamesPlayed']

dataset.head() #verification if new column is added
```

```python
dataset['Winning Percent'] #checking for NaN values in newly added column
```

```python
#replacing NaN values with 0 in win % column
dataset['Winning Percent'].fillna(0, inplace = True)
dataset['Winning Percent'] #verification replacement is successfull
```

```python
dataset.sort_values('Winning Percent',ascending = False).head() #final top 5 teams as per win% in dataset after new column addition and data cleaning
#note: winning percent is last column
```

### 6.Group teams based on their Best position and print the sum of their points for all positions (5 points)

```python

dataset['BestPosition'] = dataset['BestPosition'].astype(int) #converting type to int of the column to perform arithmetic operations
#finiding out unique best position numbers
unq = np.array(dataset.BestPosition.unique())
unq.sort()
#grouping by best positions
position = dataset.groupby(['BestPosition'])
print('Sum of the points for all positions')
print('Best Postion', 'Points\t')
#using the unique numbers calling groups and adding points for those groups
for x in unq:
    abc = position.get_group(x)
    print(x, '\t    ',abc['Points'].sum())
```
