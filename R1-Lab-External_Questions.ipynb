   2. sepal width in cm
   3. petal length in cm

### Find out the no of male & female employees in the company

*Hint: Use can use a condition to find out the no of 1s and 0s in gender and sum it*

### Find out the IQR for the column - Age 

*Hint: IQR - Interquartile range = 75th percentile - 25th percentile* 

```python
age = np.array(dataset['Age'])
iqr = np.percentile(age,75) - np.percentile(age,25)
print("IQR =", iqr)
```

### Find out the mean, median and variance for the column - DailyRate 

```python
dailr = np.array(dataset['DailyRate'])
mean = np.mean(dailr)
median = np.median(dailr)
variance = np.var(dailr)
print ("Mean = ", mean)
print ("Median = ", median)
print ("Variance = ", variance)
```

### Plot the attrition column (Since it is a categorical variable, chose the plot accordingly)

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="ticks", color_codes=True)
```

```python
#using seaborn
sns.countplot(x="Attrition", data=dataset)
plt.show()
```

```python
#using matplotlib.pyplot
yes = dataset[dataset.Attrition == 'Yes'].shape[0]
no = dataset[dataset.Attrition == 'No'].shape[0]
hei = {'Yes':yes,'No':no}

names = list(hei.keys())
values= list(hei.values())
index = np.arange(len(names))
plt.bar(index, values)
plt.xlabel('Attrition')
plt.ylabel('Count')
plt.xticks(index,names)
plt.show()
```
